# Task 2: Simulate and fit radiation shielding data

This task is based on a Physics 119 experiment, where you are seeking to measure the attentuation coefficient of some shielding with respect to a specific source of radiation. 

A simple, but effective first-order model for the behaviour of the shielding is that the radiation count rate observed in your detector can be modelled as

$$R(x) = R_0 e^{-μx} + R_b,$$

where 

* $R_0$ is the radiation count-rate observed by the detector when the radiation source is placed at a specific location/distance from the detector
* $x$ is the thickness of the shielding in between the source and detector, where this shielding has an attenuation coefficient $\mu$,
* $R_b$ is the background / environmental radiation count-rate, which is negligibly unaffected by this shielding, and 
* $R(x)$ is the count-rate measured by the detector when a thickness $x$ of shielding is present, but also includes contributions due to background/environmental radiation.

## *Task 2.2: Analyze simulated data---what is your experimentally determined attenuation coefficient,* 𝜇?

**The purpose of this task is to look more carefully at fitting parameters and fitting parameter uncertainties.
```python
### Solution

import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

def exp_model(x, a, mu, c): 
    return a * np.exp(-mu * x) + c

# Constants
Rbackground = 0.18 # counts per second; unaffected by shielding
R0 = 20  # initial count rate (counts per second) without shileding
mu = 0.5  # linear attenuation coefficient for the shielding material (1/cm). 
thicknesses = np.linspace(0, 20, 21)  # From 0cm to 10cm in 0.1cm increments

# Simulate count rate using Poisson statistics for each thickness
counts = []
t = 60*30/thicknesses.size # seconds total / number of thicknesses
for x in thicknesses:
    expected_counts = R0*t * np.exp(-mu * x) + Rbackground*t  # Reduced count rate with shielding
    count = np.random.poisson(expected_counts)  # Simulated count for this thickness using Poisson distribution
    counts.append(count)

counts = np.array(counts)
d_counts = np.sqrt(counts)
rates = counts/t
d_rates = d_counts/t 

# Do the fit
popt, pcov = optimize.curve_fit(exp_model, thicknesses, rates, 
                                sigma=d_rates, absolute_sigma=True, p0=[1000, 0.5, 11])
print("popt =",popt)
print(f"mu = {popt[1]:.4f} +/- {np.sqrt(pcov[1][1]):.4f} 1/mm")

xf = np.linspace(thicknesses.min(), thicknesses.max(), num = 50)
yf = exp_model(xf, *popt)

##

# Create main plot and residuals plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10), gridspec_kw={'height_ratios': [3, 1]})
fig.subplots_adjust(hspace=0.4)

# Plot the best fit line on main plot
ax1.plot(xf, yf, "r-", label="Fit (ignoring uncertainties)")
# Plot simulated data on main plot
ax1.errorbar(thicknesses, rates, yerr=d_rates, fmt='o', label='Simulated Count-rates')
ax1.set_xlabel('Shielding Thickness (mm)')
ax1.set_ylabel('Count Rate (cps)')
ax1.set_title('Simulated Count Rate vs. Shielding Thickness')
ax1.legend()
ax1.grid(True)

# Calculate residuals and plot on residuals plot
residuals = rates - exp_model(thicknesses, *popt)
ax2.errorbar(thicknesses, residuals, yerr=d_rates, fmt='o', color='blue')
ax2.axhline(0, color='red', linestyle='--')  # a line for zero residual
ax2.set_xlabel('Shielding Thickness (mm)')
ax2.set_ylabel('Residuals')
ax2.grid(True)
plt.show()

##
```
