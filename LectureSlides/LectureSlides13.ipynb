## 3 basic systems you can investigate
- **1 proton + 2 electrons orbiting in the same rotational directon:** Tricky to find interesting things to investigate - very sensitive to small asymmetries.
- **1 proton + 2 electrons orbiting in the opposite rotational directon:** Creates lots of different types of "orbits"
- **2 protons + 1 electron:** Largest number of different things to vary and still have stable orbits.

## You define stability
* It doesn't have to be identical repeating orbits. Many shapes precess
* It can be interesting to develelop indices or quality factors that characterize stability
