### String formatting (3 styles)
Percent-style (%) string formatting (old style)
```
print("The answer is %6.4f +/- %8.3f" % (y, dy) )
```
str.format() (new style)
```
print("The answer is {:6.4f} +/- {:8.3f}".format(y, dy))
```
f-strings
```
print(f"The answer is {y:6.4f} +/- {dy:8.3f}")
```

The `n` in `{dy:n.3f}` is the *minimum field width*
* Specifies minimum number of characters to display a value. 
* If the actual value has fewer the field will be padded with spaces to meet minimum width. 
* If the actual value has more characters, minimum field width is ignored and the full value is displayed.
* `n` does not have to be specified, e.g., `{dy:.3f}`

> **GenAI prompt:** Explain what the 6 and 8 mean in the following: print(f"The answer is {y:6.4f} +/- {dy:8.3f}")

General format: **{value:width.precisionf}**

* **width (1st number)**  minimum number of characters reserved for the output.
* **precision (number after the dot)**  how many digits to display after the decimal point.
* **f**  fixed-point notation (decimal numbers, not scientific notation).

{dy:8.3f}
* 8  reserve at least 8 characters of space.
* .3f  print 3 digits after the decimal.

Example:
```python
dy = 12.3456
print(f"{dy:8.3f}")  # -> '  12.346'
```

Two spaces are added before because the number itself is only 6 characters (12.346).

 So in summary:

* The 6 and 8 specify the minimum field widths.
* The 4 and 3 specify the number of digits after the decimal point.

### String indexing (extracting a single character from a string)
Indexing: extracting a single character from a string

````python
my_string = "Rainy days"
print(my_string[2])
print(my_string[2:3]) # Same as the one above
print(my_string[-2])
````

### String slicing (extracting a portion of a string)

**Polling Q2:** How do we print `y d`?

````python
my_string = "Rainy days"

print(my_string[3:5])  # A
print(my_string[3:6])  # B
print(my_string[4:6])  # C
print(my_string[4:7])  # D
# None of these        # E   
````

We can also use negative index slicing

````python
my_string = "Rainy days"

print(my_string[-6:])
print(my_string[:-3])
print(my_string[-6:-3])
print(my_string[-6:7])
````
